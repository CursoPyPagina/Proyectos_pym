### Dataframe con la informaci칩n simulada

Con base en lo aprendido en clases anteriores, podemos simular la informaci칩n de ventas de un d칤a.

````python
# Fecha reporte
fecha = "2025-01-04"

# Definimos listas vacias que almacenaran la informacion de las ventas
fechas = []
sucursales = []
productos = []
claves_producto = []
precios = []
cantidades_vendidas = []
totales_ticket = []

for i in range(1, 1001):
  ##############################################################################
  # 1. Generamos la informacion simulada
  ##############################################################################
  sucursal = r.choice(papelerias)
  producto = r.choice(lineas)
  clave_producto = r.choice(abcdario) + r.choice(abcdario) + r.choice(abcdario) + "-" + str(r.randint(1,9)) + str(r.randint(1,9)) + str(r.randint(1,9))
  precio = round(r.random() * r.randint(100, 10000), 2)
  cantidad_vendida = r.randint(1, 1000)
  total_ticket = precio * cantidad_vendida
  ##############################################################################
  # 2. La agregamos a nuestras listas vacias
  ##############################################################################
  fechas.append(fecha)
  sucursales.append(sucursal)
  productos.append(producto)
  claves_producto.append(clave_producto)
  precios.append(precio)
  cantidades_vendidas.append(cantidad_vendida)
  totales_ticket.append(total_ticket)

# Definimos el dataframe
import pandas as pd

# diccionario con la estructura del dataframe
diccionario_ventas_df = {
    "Fecha": fechas,
    "Sucursal": sucursales,
    "Producto": productos,
    "Clave_Producto": claves_producto,
    "Precio": precios,
    "Cantidad_Vendida": cantidades_vendidas,
    "Total_Ticket": totales_ticket
}

# Crearemos el dataframe
df_ventas = pd.DataFrame(diccionario_ventas_df)
df_ventas
````

Esto 칰ltimo nos ser치 de gran utilidad para definir nuestra primer tabla en SQL.

Antes de ello, para usar c칩digo SQL en Python deberemos de hacerlo mediante strings multil칤nea de la siguiente forma

````python
# 1. Definimos el codigo SQL (mas adelante se explica con mas detalle)
codigo_sql = """
CREATE TABLE
  VENTAS_2025 (
    Fecha TEXT,
    Sucursal TEXT,
    Producto TEXT,
    Clave_Producto TEXT,
    Precio REAL,
    Cantidad_Vendida INTEGER,
    Total_Ticket REAL
  )

"""

# 2. Hacemos la conexion a la base de datos como sigue
con = sql.connect("Prueba1.db")

# 3. Definimos un puente entre Python y SQL con el cual podremos ejecutar sentencias SQL
cursor = con.cursor()

# 4. Ejecutamos el codigo SQL
cursor.execute(codigo_sql)

# 5. Guardamos cambios
con.commit()

# 6. Cerramos la conexion
con.close()
````

En general, la estructura anterior es la estructura con la cual estaremos ejecutando sentencias SQL.

Ahora, veamos c칩mo insertar informaci칩n en nuestra tabla

````python
# Sentencia SQL
query_sql = """
INSERT INTO
  VENTAS_2025(
    Fecha,
    Sucursal,
    Producto,
    Precio,
    Cantidad_Vendida,
    Total_Ticket
    )
VALUES(
  '2024-01-05',
  'Xochimilco',
  'Libreta',
  24.00,
  7,
  168.00
)
"""

# 2. Hacemos la conexion a la base de datos como sigue
con = sql.connect("Prueba1.db")

# 3. Definimos un puente entre Python y SQL con el cual podremos ejecutar sentencias SQL
cursor = con.cursor()

# 4. Ejecutamos el codigo SQL
cursor.execute(query_sql)

# 5. Guardamos cambios
con.commit()

# 6. Cerramos la conexion
con.close()
````

````python
# 2. Hacemos la conexion a la base de datos como sigue
con = sql.connect("Prueba1.db")

cursor = con.cursor()

# Sentencia SQL, hacemos un f-string multilinea
for i in range(len(df_ventas)):
  query_sql = f"""
  INSERT INTO
    VENTAS_2025
  VALUES(
    -- Esto es un comentario en SQL
    -- Ojo con las comillas que colocamos, debido a que definimos las columnas
    -- como TEXT, es decir, como cadenas de texto
    '{df_ventas.loc[i, "Fecha"]}',
    '{df_ventas.loc[i, "Sucursal"]}',
    '{df_ventas.loc[i, "Producto"]}',
    '{df_ventas.loc[i, "Clave_Producto"]}',
    {df_ventas.loc[i, "Precio"]},
    {df_ventas.loc[i, "Cantidad_Vendida"]},
    {df_ventas.loc[i, "Total_Ticket"]}
  )
  """

  # 4. Ejecutamos el codigo SQL
  cursor.execute(query_sql)

  # 5. Guardamos cambios
  con.commit()

# 6. Cerramos la conexion
con.close()
````

````python
# Realicemos una consulta para ver la informacion
# Para ello usaremnos un comando de la libreria pandas
query_select = """
SELECT
  *
FROM
  VENTAS_2025
"""

# 2. Hacemos la conexion a la base de datos como sigue
con = sql.connect("Prueba1.db")
df = pd.read_sql(query_select, con)
con.close()

df
````
